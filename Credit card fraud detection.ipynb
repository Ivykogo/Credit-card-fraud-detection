```python
#Splitting
from sklearn.model_selection import train_test_split
X = new_df.drop(columns='Class', axis=1)
Y = new_df['Class']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Testing and training the model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression()
model.fit(X_train, Y_train)
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
#Model evaluation
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```

Accuracy on Training Data (94.84%): This shows that the model correctly predicted the labels for about 94.84% of the training data. This suggests that the model has learned quite well from the training data.

Accuracy on Test Data (93.16%): This is the accuracy on unseen data (the test set). It indicates that the model still performs quite well on data it hasn't seen before, at 93.16%.

Both accuracies are quite high, which suggests that the model is generalizing well and is not overfitting

```python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score

# Calculate Precision, Recall, F1-Score, and ROC-AUC on the test set
y_test_predictions = model.predict(X_test)

precision = precision_score(Y_test, y_test_predictions)
recall = recall_score(Y_test, y_test_predictions)
f1 = f1_score(Y_test, y_test_predictions)
roc_auc = roc_auc_score(Y_test, y_test_predictions)

print("Precision: ", precision)
print("Recall: ", recall)
print("F1-Score: ", f1)
print("ROC-AUC: ", roc_auc)
```

Precision (94.57%):
This means that 94.57% of the transactions predicted as fraud (positive class) by your model were actually fraud. It's a high precision, which is great for avoiding false positives (non-fraud transactions being classified as fraud).
Recall (91.58%):
This indicates that 91.58% of the actual fraud transactions were correctly identified by the model. (we don’t want to miss too many fraudulent transactions).
F1-Score (93.05%):
The F1-score is a balance between precision and recall. Since it's fairly close to both precision and recall, it suggests a good balance between identifying fraud cases and avoiding false alarms.

```python
from sklearn.model_selection import GridSearchCV

# Set up the parameter grid
param_grid = {
    'C': [0.1, 1, 10],  # Regularization strength
    'solver': ['lbfgs', 'liblinear']
}

# Perform Grid Search with cross-validation
grid_search = GridSearchCV(LogisticRegression(), param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, Y_train)

# Output the best parameters
print("Best Parameters: ", grid_search.best_params_)
```

```python
# Retrain the model using the best parameters
model = LogisticRegression(C=0.1, solver='lbfgs')
model.fit(X_train, Y_train)
```
